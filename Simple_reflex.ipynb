```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import time
```

```python
environment={
    "Room1":"Clean",
    "Room2":"Dirty",
    "Room3":"Clean",
    "Room4":"Clean"

}
```

```python
room_position={
    "Room1":[0,1],
    "Room2":[1,1],
    "Room3":[0,0],
    "Room4":[1,0]
}
```

```python
rooms=list(environment.keys())
agent_index=0
```

```python
def reflex_agent(state):
  if state=="Dirty":
    return "Dirty"
  else:
    return "Move"
```

```python
def draw_environment(env, agent_pos, step):
    fig, ax = plt.subplots()
    ax.set_xlim(0, 2)
    ax.set_ylim(0, 2)
    ax.set_xticks([0, 1])
    ax.set_yticks([0, 1])
    ax.set_title(f"Step {step}-Agent in {rooms[agent_pos]}")
    ax.set_xticklabels(["Room1", "Room2"])
    ax.set_yticklabels(["Room3", "Room4"])
    for room, pos in room_position.items():
        x, y = pos
        color = 'red' if env[room] == "Dirty" else 'green'
        rect = patches.Rectangle((x, y), 1, 1, linewidth=1, edgecolor='black', facecolor=color)
        ax.add_patch(rect)
        ax.text(x + 0.5, y + 0.5, room, fontsize=10, ha='center', va='center')

    agent_x, agent_y = room_position[rooms[agent_pos]]
    agent_patch = patches.Rectangle((agent_x, agent_y), 1, 1, linewidth=2, edgecolor='blue', facecolor='white')
    ax.add_patch(agent_patch)
    plt.show()
```

```python
draw_environment(environment, agent_index, 0)
```

```python
plt.pause(1)
plt.close()
```

```python
plt.ion()
steps=8
```

```python
for step in range(steps):
  current_room=rooms[agent_index]
  state=environment[current_room]
  action=reflex_agent(state)
  draw_environment(environment, agent_index, step+1)
  if action=="Clean":
    environment[current_room]="Clean"
  else:
    agent_index=(agent_index+1)%len(rooms)
plt.ioff()
print("Simulation completed")
```
